```python
import panadas as pd

df = pd.read_csv(
    "mumbai_synthetic_congestion_dataset_2y.csv",
    parse_dates=["timestamp"]
)

df = df.sort_values(["road_id", "timestamp"])
```

```python
LAGS = [1, 2, 3, 6, 24]

for lag in LAGS:
    df[f"congestion_lag_{lag}"] = (
        df.groupby("road_id")["congestion_level"]
          .shift(lag)
    )
```

```python
df["speed_lag_1"] = df.groupby("road_id")["avg_speed_kph"].shift(1)
df["delay_lag_1"] = df.groupby("road_id")["delay_ratio"].shift(1)
```

```python
HORIZONS = [1, 2, 6, 24]

for h in HORIZONS:
    df[f"target_congestion_t_plus_{h}h"] = (
        df.groupby("road_id")["congestion_level"]
          .shift(-h)
    )
```

```python
FEATURES = [
    # Traffic memory
    "congestion_lag_1", "congestion_lag_2", "congestion_lag_3",
    "congestion_lag_6", "congestion_lag_24",
    "speed_lag_1", "delay_lag_1",

    # Time
    "hour", "day_of_week", "is_weekend", "month",

    # Road identity
    "road_class",

    # Risk priors
    "accident_hotspot_score",
    "recent_incident_count",
    "enforcement_violation_pattern",
    "long_term_risk_prior"
]

TARGETS = [f"target_congestion_t_plus_{h}h" for h in HORIZONS]
```

```python
model_df = df[
    FEATURES + TARGETS + ["road_id", "timestamp"]
].dropna()
```

```python
model_df = model_df.sort_values("timestamp")
```
